### Initialize variables

```python
# Initialize variables used in the program 
MSS_Name = '0'
MSS_Date = '0'
MSS_Time = '0'
MSS_Pars = [MSS_Name, MSS_Date, MSS_Time]

LAC_str = '0'
LAC_Name = '0'
ZC_0_Stat = ZC_1_Stat = ZC_2_Stat = ZC_3_Stat = ZC_4_Stat = ZC_5_Stat = '0'

LAC_Pars = [LAC_str, LAC_Name, ZC_0_Stat, ZC_1_Stat, ZC_2_Stat, ZC_3_Stat, ZC_4_Stat, ZC_5_Stat]

ZC_0_Num = ZC_1_Num = ZC_2_Num = ZC_3_Num = ZC_4_Num = ZC_5_Num = 'N/A'
ZC_0_Name = ZC_1_Name = ZC_2_Name = ZC_3_Name = ZC_4_Name = ZC_5_Name = "N/A"
ZC_0_ROAM = ZC_1_ROAM = ZC_2_ROAM = ZC_3_ROAM = ZC_4_ROAM = ZC_5_ROAM = "N/A"
ZC_0_IN_MM = ZC_1_IN_MM = ZC_2_IN_MM = ZC_3_IN_MM = ZC_4_IN_MM = ZC_5_IN_MM = "N/A"

ZC_Stat = [ ZC_0_Stat, ZC_1_Stat, ZC_2_Stat, ZC_3_Stat, ZC_4_Stat, ZC_5_Stat]
ZC_0_Pars = [ZC_0_Num, ZC_0_Name, ZC_0_ROAM, ZC_0_IN_MM]
ZC_1_Pars = [ZC_1_Num, ZC_1_Name, ZC_1_ROAM, ZC_1_IN_MM]
ZC_2_Pars = [ZC_2_Num, ZC_2_Name, ZC_2_ROAM, ZC_2_IN_MM]
ZC_3_Pars = [ZC_3_Num, ZC_3_Name, ZC_3_ROAM, ZC_3_IN_MM]
ZC_4_Pars = [ZC_4_Num, ZC_4_Name, ZC_4_ROAM, ZC_4_IN_MM]
ZC_5_Pars = [ZC_5_Num, ZC_5_Name, ZC_5_ROAM, ZC_5_IN_MM]

index = 0
position = 0
LAC_count = 0
```

```python
# Create & Initialize & Set the pd.DataFrame
col_names = ['MSS','Date','Time', 'LAC_str','LAC_NAME', 'Claro_Roam', 'Vivo_Roam', 'Oi_Roam', 'Tim_Roam', 'CTBC_Roam', 'Next_Roam']
df_MSS_LACs = pd.DataFrame(columns=col_names)
df_MSS_LACs.loc[0] = [MSS_Name, MSS_Date, MSS_Time, LAC_str, LAC_Name, ZC_0_Stat, ZC_1_Stat, ZC_2_Stat, ZC_3_Stat, ZC_4_Stat, ZC_5_Stat]
```

```python
# Especify the path of the log file
path_to_file = 'C:/Users/Alvaro/OneDrive/My Files/_Scripts_Models/Commands Generators Logs/20220204 - ZEKO - All_MSS.txt'
```

### Load log file

```python
# Load the log file
log_file = NOK_ERI_Toolkit.open_file(path_to_file)
```

### Execute & Create pd.DataFrame & Pring (Main Function)

```python
# Run the main program (txt => pd.DataFrame)
while (index < len(log_file)-1):
    index, position, MSS_Pars = NOK_ERI_Toolkit.get_mss_pars(log_file, index, 'ZEKO:LAC')
    while (index < len(log_file)-1) and (not position == 999):
        index, position, LAC_Pars = NOK_ERI_Toolkit.get_zeko_lac_pars(log_file, index)
        if position == 999:                                         # if 'COMMAND EXECUTED'
            index = index + 1                                       # Increase index to read next line
        else:
            new_LAC = MSS_Pars + LAC_Pars
            df_MSS_LACs.loc[LAC_count] = new_LAC
            LAC_count = LAC_count + 1
```
