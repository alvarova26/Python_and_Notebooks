```python
# Import modules (NOK_Toolkit) & others that will be needed along the program
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import NOK_Toolkit as NOK_Toolkit

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

## Main Program

### Initialize variables

```python
# Initialize variables used in the program 
MSS_Name = '0'
MSS_Date = '0'
MSS_Time = '0'
MSS_Pars = [MSS_Name, MSS_Date, MSS_Time]

LAC_str = '0'
LAC_Name = '0'
ZC_0_Stat = ZC_1_Stat = ZC_2_Stat = ZC_3_Stat = ZC_4_Stat = ZC_5_Stat = '0'

LAC_Pars = [LAC_str, LAC_Name, ZC_0_Stat, ZC_1_Stat, ZC_2_Stat, ZC_3_Stat, ZC_4_Stat, ZC_5_Stat]

index = 0
position = 0
LAC_count = 0
```

```python
# Create & Initialize & Set the pd.DataFrame
col_names = ['MSS','Date','Time', 'LAC_str','LAC_NAME', 'Claro_Roam', 'Vivo_Roam', 'Oi_Roam', 'Tim_Roam', 'CTBC_Roam', 'Next_Roam']
df_MSS_LACs = pd.DataFrame(columns=col_names)
df_MSS_LACs.loc[0] = [MSS_Name, MSS_Date, MSS_Time, LAC_str, LAC_Name, ZC_0_Stat, ZC_1_Stat, ZC_2_Stat, ZC_3_Stat, ZC_4_Stat, ZC_5_Stat]
df_MSS_LACs
```

```python
# Especify the path of the log file
path_to_file = 'C:/Users/Alvaro/OneDrive/My Files/_Scripts_Models/Commands Generators Logs/20220204 - MSSBAxERICC - MGNRP.txt'
```

### Load log file

```python
# Load the log file
log_file = NOK_Toolkit.open_file(path_to_file)
```

### Execute & Create pd.DataFrame & Pring (Main Function)

```python
# Run the main program (txt => pd.DataFrame)
while (index < len(log_file)-1):
    index, position, MSS_Pars = NOK_Toolkit.get_mss_pars(log_file, index,'eaw')
    while (index < len(log_file)-1) and (not position == 999):
        index, position, LAC_Pars = NOK_Toolkit.get_mgnrp_lac_pars(log_file,index)
        if position == 999:                                         # if 'END'
            index = index + 1                                       # Increase index to read next line
        else:
            new_LAC = MSS_Pars + LAC_Pars
            df_MSS_LACs.loc[LAC_count] = new_LAC
            LAC_count = LAC_count + 1
df_MSS_LACs
```

### Analyze Data

```python
# Working on it...

drop_col = []
df_MSS_LAIs = df_MSS_LACs.drop(columns=drop_col)

# Very efficient way to split a string column to two or more columns based on a pattern
df_MSS_LAIs[['MCC','MNC', 'LAC_int']] =  df_MSS_LAIs['LAC_NAME'].str.split('-', 2, expand=True)
df_MSS_LAIs['LAI'] =  df_MSS_LAIs['LAC_NAME']
df_MSS_LAIs['NE-LAI'] =  df_MSS_LAIs['MSS'] + '-' + df_MSS_LAIs['LAI']
df_MSS_LAIs
```
