## Import modules

```python
import pandas as pd
import numpy as np
import re as re
from matplotlib import pyplot as plt
import seaborn as sns
```

## Load data and clean it up if necessary

```python
xls = pd.ExcelFile('PROJETO ROAMING 30K - Novas Liberações Rev1 - BASE.xlsx') #Read excel file
df1 = pd.read_excel(xls, 'LAC')                                               #Read specific sheet 
df1 = df1.drop_duplicates(subset=['MUNICIPIO'])                                     #Drop duplicates by specific column
df1                                                                           #View data
```

```python
df2 = df1['MSS']                                 #Get a specific column
df2 = df2.drop_duplicates()                      #Drop duplicates
df2 = df2.reset_index().drop('index', axis=1)    #Reset index and drop the created column 'index'
df2 = df2.sort_values('MSS', ascending=True)     #Sort values by column 'MSS'
df2 = df2.reset_index().drop('index', axis=1)    #Reset index and drop the created column 'index'
df2                                              #View data
```

```python
def print_commands(df, mss):
    for municipio in range(len(df)):
      MSS = str(df.MSS[municipio])
      if MSS == mss:
        LAI = '724-05-' + str(df.LAC[municipio])
        CN = str(df.LAC[municipio])[-2:]
        Municipio = str(df.MUNICIPIO[municipio])
        RNC = str(df.RNC[municipio])
        NRRG = '7&8&9&12&13'

        if 'VIVO' not in str(df.R30K[municipio]):
          NRRG = '6&'+ NRRG

        if 'OI' not in str(df.R30K[municipio]):
          NRRG = '2&'+ NRRG

        if 'TIM' not in str(df.R30K[municipio]):
          NRRG = '1&'+ NRRG

        print('!' + Municipio)
        print('MGLAI:LAI=' + LAI + ';')
        print('MGLXI:LATA=' + CN + ',LAI=' + LAI + ';')
        print('MGNRI:LAI=' + LAI + ',NRRG=' + NRRG + ';')
        print('MGMAI:LAI=' + LAI + ',RNC=' + RNC + ';' + '\n')
```

## PRINT CHANGING COMMANDs

```python
for i in range(len(df2)):
  MSS = df2.MSS[i]

  if ((MSS == 'MSSBA3') or (MSS == 'MSSBA7') or (MSS == 'MSSBA8')):
    print('! ' + MSS)
    print('!--------------------------------------------------------------------------------------')
    print('!--------------------------------------------------------------------------------------')
    print_commands(df1, MSS)
```
